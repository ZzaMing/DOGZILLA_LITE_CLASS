```python
import ipywidgets as widgets
from IPython.display import display
import sys

# 创建输出区域
output_area = widgets.Output(layout={'width': '100%', 'height': '300px', 'border': '1px solid #ccc'})

# 显示输出区域
display(output_area)

# 保存原始标准输出和错误输出
original_stdout = sys.stdout
original_stderr = sys.stderr

# 自定义输出重定向器
class SimpleRedirector:
    def __init__(self, output_area):
        self.output_area = output_area
        
    def write(self, text):
        with self.output_area:
            # 使用IPython的显示系统来输出文本
            print(text, end='')
            
    def flush(self):
        pass

# 创建重定向器实例并应用到标准输出和错误输出
redirector = SimpleRedirector(output_area)
sys.stdout = redirector
sys.stderr = redirector

# 演示
print("所有输出将重定向到上面的输出框")
print("这是一条普通消息")
print("这是一条错误消息", file=sys.stderr)

for i in range(10):
    print(f"计数: {i}")
```
